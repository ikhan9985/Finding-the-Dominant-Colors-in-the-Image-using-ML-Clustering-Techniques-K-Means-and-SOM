```python
from minisom import MiniSom
```

```python
img = plt.imread("/content/gdrive/MyDrive/A1/dominant-colors.jpg")
```

```python
data=img.reshape(img.shape[0]*img.shape[1],3)
data.shape
```

```python
scaler2 = StandardScaler()

scaler2.fit(data)
```

```python
scaled_data = scaler2.transform(data)
scaled_data.shape
```

```python
Color_Model = MiniSom(3, 3, 3, sigma=0.1, learning_rate=0.2)
Color_Model.train(scaled_data, 10000,verbose=True,)
```

```python
scaled_data_qnt=Color_Model.quantization(scaled_data)

scaled_data_qnt.shape
```

```python
data_qnt=scaler2.inverse_transform(scaled_data_qnt)
data
```

```python
img_qnt=data_qnt.reshape(img.shape[0],img.shape[1],3)
```

```python
img_qnt.shape
```

```python
from google.colab.patches import cv2_imshow
from matplotlib.pyplot import imread,imshow,figure,show,subplot,title
from numpy import reshape,flipud,unravel_index,zeros
```

```python
print("ORIGINAL IMAGE")
cv2_imshow(img)
print("\n")
print("RESULT IMAGE")
cv2_imshow(img_qnt)
```

```python
print('building new image...')
clustered = zeros(img.shape)
for i,q in enumerate(data_qnt): # place the quantized values into a new image
    clustered[unravel_index(i,shape=(img.shape[0],img.shape[1]))] = q
print('done.')
clustered
```

```python
cv2_imshow(clustered)
```

```python
title('learned colors')
imshow(flipud(Color_Model.get_weights()),interpolation='none')
show()
```
